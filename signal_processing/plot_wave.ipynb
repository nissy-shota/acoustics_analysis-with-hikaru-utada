```python
import os
import librosa
```

```python
ACOUSTIC_DIR = '/content/drive/MyDrive/acousitics_analysis'
ALBUM_NAME = 'Fantome'
MUSIC_NAME  = 'oreno-kanojyo'
WAV_EXT = 'wav'

music_file_name = MUSIC_NAME + '.' + WAV_EXT
```

```python
music_file_path = os.path.join(ACOUSTIC_DIR, ALBUM_NAME, music_file_name)
```

```python
import librosa.display

sr=44100
music, sr_ = librosa.load(music_file_path, sr=sr)

print("play time : {} [sec]".format(len(music) / sr))
librosa.display.waveplot(music, sr=sr)
```
