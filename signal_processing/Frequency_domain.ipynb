```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
ACOUSTIC_DIR = '/content/drive/MyDrive/acousitics_analysis'
ALBUM_NAME = 'Fantome'
MUSIC_NAME  = 'oreno-kanojyo'
WAV_EXT = 'wav'

music_file_name = MUSIC_NAME + '.' + WAV_EXT
```

```python
music_file_path = os.path.join(ACOUSTIC_DIR, ALBUM_NAME, music_file_name)
```

```python
music, sr = librosa.load(music_file_path)
```

```python
D = np.abs(librosa.stft(music, n_fft=2048, hop_length=512))
print(D.shape)

plt.figure(figsize=(16, 6))
plt.plot(D)
plt.grid()
plt.show()
```
